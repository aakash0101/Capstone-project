## Literacy Rate According to Census from [1951-2011]

```python
# Selecting columns of interest
years = ['1951', '1961', '1971', '1981', '1991', '2001', '2011']
columns_to_plot = [f'Literacy_rate_{year}' for year in years]

plt.figure(figsize=(25, 10))

# Iterating through columns to plot each line
for column in columns_to_plot:
    plt.plot(df['State&Uts'], df[column], marker='o', label=column)

plt.xlabel('State & Uts', fontsize=25)
plt.ylabel('Literacy Rate Over The Years', fontsize=25)
plt.title('Literacy Rate According To Census')
plt.xticks(fontsize=15, rotation=90)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

The literacy rate in India has shown remarkable improvement from the 1951 census to the 2011 census, reflecting the nation's evolving educational landscape.

1951 Census: The literacy rate in India was approximately 18.33% in 1951, highlighting widespread illiteracy and limited educational access.

1971 Census: By 1971, the literacy rate had increased to about 29.45%, attributed to government-led educational initiatives and infrastructure development.

1991 Census: In 1991, the literacy rate rose to around 52.21%, driven by expanded educational opportunities and awareness programs.

2001 Census: By 2001, the literacy rate reached approximately 64.84%, with a focus on universal primary education and gender equality in schooling.

2011 Census: In 2011, India achieved a literacy rate of about 74.04%, marking significant progress due to continued government investment and enhanced educational awareness.

While these gains are notable, challenges like regional disparities and quality of education persist. The Indian government remains committed to improving literacy through targeted policies and initiatives.

# Types of Crime In INDIA

```python
plt.figure(figsize=(26, 10))
plt.title('Types of Crimes and Average NSO Literacy Rate (2017)', fontsize=25)

# Define crime columns to plot
crime_columns = ['Violent_Crimes', 'Murder', 'Rape', 'Kidnapping', 
                 'Crimes_Against_Children', 'Extortion', 'Robbery', 
                 'Hit_&_Run', 'Drugs_Trafficking', 'Illegal_Arms']

# Plotting crimes
for crime in crime_columns:
    plt.plot(df['State&Uts'], df[crime], marker='o', label=crime)

# Plotting average NSO literacy rate
plt.plot(df['State&Uts'], df['Average_NSO_literacy_rate_2017'], 
         marker='o', label='Average NSO Literacy Rate (2017)', linewidth=5)

plt.xlabel('State & Uts', fontsize=25)
plt.ylabel('Number of Crimes', fontsize=25)
plt.xticks(fontsize=18, rotation=90)
plt.legend(fontsize='large')
plt.grid(True)
plt.tight_layout()
plt.show()
```

-Crime Against Children is notably high across several states, including Andaman & Nicobar, Delhi, Sikkim, Haryana, Chandigarh, Madhya Pradesh, Chhattisgarh, Maharashtra, and Lakshadweep, consistently above average compared to other states.

-Cases involving illegal arms are particularly prevalent in Uttar Pradesh and Uttarakhand.

-Violent crime rates are prominently high across all states, with Assam, Delhi, Odisha, and Haryana showing notably elevated rates.

-Delhi exhibits the highest incidence of kidnapping among all states.
